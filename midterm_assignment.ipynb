### 問題1

$\{1, 2, \dots, n\}$ の番号が付いた$n$個のボールの中から $m$ 個のボールを選び, 選んだボールの中で最も大きい番号に着目して場合分けをする.  
選んだ $m$ 個のボールの中で最も大きい番号を $j$ とする.  

##### 1. $j$ がとりうる値の範囲
選ばれる $m$ 個のボールのうち最も大きい番号 $j$ は, 最小で $m$ ($\{1, 2, \dots, m\}$ を選んだ場合) であり, 最大で $n$ となる. したがって、$m \le j \le n$ となる.

##### 2. 各場合における場合の数
最も大きい番号が $j$ であると固定した場合, 残りの $m-1$ 個のボールは $j$ より小さい番号, つまり $\{1, 2, \dots, j-1\}$ の $j-1$ 個の中から選ぶことになる. その場合の数は $_{j-1} C_{m-1}$ 通りである.

##### 3. 総和
$j$ がとりうる全ての可能性について和をとることで、全体の総数 $_n C_m$ が求められる.  
$$_n C_m = \sum_{j=m}^{n} {_{j-1} C_{m-1}}.$$

$k = j-1$ とおくと,  
$$_n C_m = \sum_{k=m-1}^{n-1} {_k C_{m-1}}.$$

### 問題4

確率過程 $X_t = W_t^2 - t$ が, $(\mathcal{F}_t)_{t \ge 0}$ に関するマルチンゲールであるためには,

1.  **可測性**: $X_t$ は $\mathcal{F}_t$ 可測.
2.  **可積分性**: $E[|X_t|] < \infty$.
3.  **マルチンゲール性**: 任意の $s < t$ に対して, $E[X_t | \mathcal{F}_s] = X_s$.

を満たせばよい.

##### 1. **可測性**

ブラウン運動 $(W_t)$ は $(\mathcal{F}_t)$ に適合しているため, $W_t$ は $\mathcal{F}_t$ 可測. 可測な確率変数の可測関数である $W_t^2$ や, 時間 $t$ も $\mathcal{F}_t$ 可測である. したがって, これらの差である $X_t = W_t^2 - t$ も $\mathcal{F}_t$ 可測である.

##### 2. **可積分性**

$E[|X_t|] < \infty$ を示す. 三角不等式より,
$$
E[|X_t|] = E[|W_t^2 - t|] \le E[|W_t^2|] + |t| = E[W_t^2] + t.
$$
$W_t \sim N(0, t)$, $Var(W_t) = E[W_t^2] - (E[W_t])^2$ より,
$$
t = E[W_t^2] - 0^2 \implies E[W_t^2] = t.
$$
よって,
$$
E[|X_t|] \le t + t = 2t < \infty.
$$


##### 3. **マルチンゲール性**

$s < t$ なる任意の $s, t$ に対して, $E[X_t | \mathcal{F}_s] = X_s$ を示す.
$$
E[W_t^2 - t | \mathcal{F}_s] = E[W_t^2 | \mathcal{F}_s] - E[t | \mathcal{F}_s] = E[W_t^2 | \mathcal{F}_s] - t.
$$
ここで $W_t$ を $W_s$ と増分 $W_t - W_s$ に分解する.
$$
W_t^2 = (W_s + (W_t - W_s))^2 = W_s^2 + 2W_s(W_t - W_s) + (W_t - W_s)^2.
$$
この式の条件付き期待値をとると,
$$
E[W_t^2 | \mathcal{F}_s] = E[W_s^2 | \mathcal{F}_s] + E[2W_s(W_t - W_s) | \mathcal{F}_s] + E[(W_t - W_s)^2 | \mathcal{F}_s].
$$

* $W_s^2$ は $\mathcal{F}_s$ 可測なので, $E[W_s^2 | \mathcal{F}_s] = W_s^2$.
* $2W_s$ は $\mathcal{F}_s$ 可測であり, 増分 $W_t - W_s$ は $\mathcal{F}_s$ と独立なので,
    $E[2W_s(W_t - W_s) | \mathcal{F}_s] = 2W_s E[W_t - W_s | \mathcal{F}_s] = 2W_s E[W_t - W_s] = 2W_s \cdot 0 = 0$.
* 増分 $(W_t - W_s)^2$ は $\mathcal{F}_s$ と独立なので,
    $E[(W_t - W_s)^2 | \mathcal{F}_s] = E[(W_t - W_s)^2] = Var(W_t - W_s) = t-s$.

以上をまとめると,
$$
E[W_t^2 | \mathcal{F}_s] = W_s^2 + 0 + (t - s) = W_s^2 + t - s.
$$
したがって,
$$
E[W_t^2 - t | \mathcal{F}_s] = (W_s^2 + t - s) - t = W_s^2 - s = X_s.
$$


以上より, 確率過程 $(W_t^2 - t)_{t \ge 0}$ は $(\mathcal{F}_t)$ マルチンゲールである.

#### (1)

```r
lbop_p_fixed <- function(S, K, r, sig, T, n, M){
    C <- 0
    dt <- T/M
    for (i in 1:n){
        St <- S
        S_min <- S
        x <- rnorm(M, 0, 1)
        for (j in 1:M){
            St <- St * exp((r - 0.5 * sig^2) * dt + sig * sqrt(dt) * x[j])
            if (St < S_min){
                S_min <- St
            }
        }
        C <- C + max(K - S_min, 0) 
    }
    C <- exp(-r * T) * C / n
    return(c("Fixed ルックバックプット価格" = C))
}
```

```r
lbop_p_fixed(S=100, K=110, r=0.01, sig=0.2, T=1, n=10000, M=100)
```

#### (3)

##### Floating Strike型ルックバックオプション価格

```r
lbop_p_floating <- function(S, K, r, sig, T, n, M){
    C <- 0
    dt <- T / M
    for (i in 1:n){
        St <- S 
        S_max <- S 
        x <- rnorm(M, 0, 1) 
        for (j in 1:M){
            St <- St * exp((r - 0.5 * sig^2) * dt + sig * sqrt(dt) * x[j])
            if (St > S_max){
                S_max <- St
            }
        }
        C <- C + max(S_max - St, 0) 
    }
    C <- exp(-r * T) * C / n
    return(c("Floating ルックバック価格" = C))
}
```

```r
lbop_p_floating(S=100, K=110, r=0.01, sig=0.2, T=1, n=10000, M=10)
```

##### Fixed型とFloating型の違い

* 利益を大きく得るシナリオ  
Fixed型: 権利行使価格 ($K$) よりも大きく下落した場合.  
Floating型: 最高値に達した後、満期時にその最高値より下落した場合.  

* リスクヘッジ  
Fixed型: 権利行使価格 ($K$) より期間中の最安値が大きいとペイオフが0となる.  
Floating型: 満期時の価格=最高値 を除き必ず正のペイオフを得る.  

* 低ボラティリティのとき($S_{min} = S_{max} = S$ を仮定)  
Fixed型: 権利行使価格 ($K$) > 原資産価格 ($S$) であれば $K-S$ のペイオフを得られる.  
Floating型: ペイオフは0となる.  

* 高ボラティリティのとき  
Fixed型: $S_{min}$に下振れを期待できるが, ($K$) は固定.    
Floating型: $S_{T}$ に下振れ, $S_{max}$ に上振れを期待できる.  
どちらの型がより多くのペイオフを得ることができるかは状況によるが, Floating型の方がペイオフの上振れを期待できる.

```r
# sig の範囲設定
sig_range <- seq(0.01, 1.5, by = 0.01)

# Fixed ルックバックプット
fixed_values_sig <- sapply(sig_range, function(sig) {
  lbop_p_fixed(S_base, K_base, r_base, sig, T_base, n_base, M_base)
})
plot(sig_range, fixed_values_sig, type = "l", col = "blue",
     xlab = "ボラティリティ (sig)", ylab = "Fixed ルックバックプット価格",
     main = "ボラティリティ (sig) とFixedルックバックプット価格の関係")
grid()


# Floating ルックバックプット
floating_values_sig <- sapply(sig_range, function(sig) {
  lbop_p_floating(S_base, K_base, r_base, sig, T_base, n_base, M_base)
})
plot(sig_range, floating_values_sig, type = "l", col = "blue",
     xlab = "ボラティリティ (sig)", ylab = "Floating ルックバックプット価格",
     main = "ボラティリティ (sig) とFloatingルックバックプット価格の関係")
grid()

```

ルックバックプットオプション価格について,  
* 低ボラティリティではFixed型 > Floating型  
* 高ボラティリティではFixed型 < Floating型

#### 1.

\begin{align*}
S\phi(d_1) &= Ke^{-rT}\phi(d_2) の導出.\\
\\

&= \ln\left( e^{\frac{1}{2}(d_1^2 - d_2^2)} \right) \\

&= \frac{1}{2}(d_1 - d_2)(d_1 + d_2) && *d_2 = d_1 - \sigma\sqrt{T} \\
&= \frac{1}{2}(\sigma\sqrt{T})(2d_1 - \sigma\sqrt{T}) \\
&= d_1\sigma\sqrt{T} - \frac{1}{2}\sigma^2T && *d_1\sigma\sqrt{T} = \ln(S/K) + (r + \frac{1}{2}\sigma^2)T \\
&= \left(\ln\frac{S}{K} + (r + \frac{1}{2}\sigma^2)T\right) - \frac{1}{2}\sigma^2T \\
&= \ln\frac{S}{K} + rT \\
\\
よって,

\\
整理して,　
K\phi(d_2) &= Se^{rT}\phi(d_1) \\
S\phi(d_1) &= Ke^{-rT}\phi(d_2)
\end{align*}

\begin{align*}

\end{align*}

#### 2.

```r
# 1. デルタ (dC/dS)
get_delta <- function(S, K, r, sig, T) {
  d1 <- (log(S / K) + (r + 0.5 * sig^2) * T) / (sig * sqrt(T))
  delta <- pnorm(d1)
  return(delta)
}

# 2. Kに関する微分 (dC/dK)
get_dC_dK <- function(S, K, r, sig, T) {
  d1 <- (log(S / K) + (r + 0.5 * sig^2) * T) / (sig * sqrt(T))
  d2 <- d1 - sig * sqrt(T)
  dC_dK <- -exp(-r * T) * pnorm(d2)
  return(dC_dK)
}

# 3. ロー (dC/dr)
get_rho <- function(S, K, r, sig, T) {
  d1 <- (log(S / K) + (r + 0.5 * sig^2) * T) / (sig * sqrt(T))
  d2 <- d1 - sig * sqrt(T)
  rho <- K * T * exp(-r * T) * pnorm(d2)
  return(rho)
}

# 4. ベガ (dC/dsig)
get_vega <- function(S, K, r, sig, T) {
  d1 <- (log(S / K) + (r + 0.5 * sig^2) * T) / (sig * sqrt(T))
  vega <- S * dnorm(d1) * sqrt(T)
  return(vega)
}

# 5. セータ (dC/dt)
get_theta <- function(S, K, r, sig, T) {
  d1 <- (log(S / K) + (r + 0.5 * sig^2) * T) / (sig * sqrt(T))
  d2 <- d1 - sig * sqrt(T)
  theta <- -(S * dnorm(d1) * sig) / (2 * sqrt(T)) - r * K * exp(-r * T) * pnorm(d2)
  # 1年(365日)当たりの値に変換
  return(theta / 365)
}
```

```r
# 基準となるパラメータを設定
S_base <- 100
K_base <- 110
r_base <- 0.01
sig_base <- 0.2
T_base <- 1
```

---

```r
# 1. デルタ vs 原資産価格(S)
S_range <- seq(60, 160, by = 1)
delta_values <- get_delta(S_range, K_base, r_base, sig_base, T_base)
plot(S_range, delta_values, type = "l", col = "blue",
     main = "デルタ(dC/dS)の推移", xlab = "原資産価格 (S)", ylab = "デルタ (dC/dS)")
abline(v = K_base, col="red", lty=2) # Kのライン
text(K_base, min(delta_values), "K=110", pos=4, col="red")
grid()
```

デルタは $0$ (ディープ・アウト・オブ・ザ・マネー) から $1$ (ディープ・イン・ザ・マネー) の値をとる  

---

```r
# 2. dC/dK vs 権利行使価格(K)
K_range <- seq(60, 160, by = 1)
dC_dK_values <- get_dC_dK(S_base, K_range, r_base, sig_base, T_base)
plot(K_range, dC_dK_values, type = "l", col = "blue",
     main = "Kに関する偏微分の推移", xlab = "権利行使価格 (K)", ylab = "dC/dK")
abline(v = S_base, col="red", lty=2) # Sのライン
text(S_base, min(dC_dK_values), "S=100", pos=4, col="red")
grid()
```

$\frac{\partial C}{\partial K} $ は $-1$ (ディープ・イン・ザ・マネー) から $0$ (ディープ・アウト・オブ・ザ・マネー) の値をとる. 

---

```r
# 3. ロー vs 無リスク利子率(r)
r_range <- seq(0.01, 1, by = 0.01)
rho_values <- get_rho(S_base, K_base, r_range, sig_base, T_base)
plot(r_range, rho_values, type = "l", col = "blue",
     main = "ロー(dC/dr)の推移", xlab = "無リスク利子率(r)", ylab = "ロー(dC/dr)")
grid()
```

* $r>0$ のとき, ローは正  
→金利が上がると $S_T$ の上昇, すなわちペイオフの上昇が期待される.
* $r$が比較的小さいとき, ローは増加  
→ロー $(KTe^{-rT}\Phi(d_2))$ における $e^{-rT}$ の減少割合よりも $\Phi{(d_2)}$ の増加割合の方が大きい.
* $r$が比較的大きいとき, ローは減少  
→ローの定義式における $e^{-rT}$ の減少割合よりも $\Phi{(d_2)}$ の増加割合の方が小さい.
---

```r
# 4. ベガ vs ボラティリティ(sig)
sig_range <- seq(0.05, 1, by = 0.005)
vega_values <- get_vega(S_base, K_base, r_base, sig_range, T_base)
plot(sig_range, vega_values, type = "l", col = "blue",
     main = "ベガ(dC/dσ)の推移", xlab = "ボラティリティ (σ)", ylab = "ベガ(dC/dσ)")
grid()
```

ベガ $(S\phi(d_1)\sqrt{T})$ における $\phi(d_1)$ の$\sigma$ に対する増加率に応じてグラフの形が変化している.

---

```r
# 5. セータ vs 満期までの期間(T)
T_range <- seq(0.01, 10, by = 0.01)
theta_values <- get_theta(S_base, K_base, r_base, sig_base, T_range)
plot(T_range, theta_values, type = "l", col = "blue",
     main = "セータ(dC/dt)の推移", xlab = "満期までの期間 (T)", ylab = "セータ(dC/dt)")
grid()
```

* セータは常に負であるが, これは時間的価値の減少によるものであるといえる.
* グラフが $T = 0$ 付近で急激に減少しているが, これはセータ $(-\frac{S\phi(d_1)\sigma}{2\sqrt{T}} - rKe^{-rT}\Phi(d_2))$ における $-\frac{S\phi(d_1)\sigma}{2\sqrt{T}}$ の分母に $T$が含まれているためであるといえる.
